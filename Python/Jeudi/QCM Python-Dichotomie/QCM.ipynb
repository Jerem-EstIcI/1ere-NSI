```python
def maxi(L):
    dernier_indice=len(L)-1
    valeur_max=L[0]
    for i in range(1,dernier_indice):
        if L[i]>valeur_max:
            valeur_max=L[i]
    return valeur_max

print(maxi([1,2,3,4]))
print(maxi([4,3,2,1]))
print(maxi([1,3,3,2]))
print(maxi([1,1,1,1]))
```

## Q8

```python
def dicho(e,t):
    """
    >>> dicho(29,[14,22,29,35,44,50,56,76])
    True
    >>> dicho(61,[14,22,29,35,44,50,56,76])
    False
    """
    i=0
    j=len(t)-1
    while i<=j:
        m=(i+j)//2
        if t[m]==e:
            return True
        elif e<t[m]:
            j=m-1
        else:
            i=m+1
    return False

print(dicho(29,[14,22,29,35,44,50,56,76]))
print(dicho(61,[14,22,29,35,44,50,56,76]))
```

### Utiliser des noms de variables plus pertinents
Rcrire la fonction dicho avec des noms de paramtres et de variables plus **explicites**.

```python
def dicho(element,tableau):
    """
    >>> dicho(29,[14,22,29,35,44,50,56,76])
    True
    >>> dicho(61,[14,22,29,35,44,50,56,76])
    False
    """
    debut=0
    fin=len(tableau)-1
    
    while debut<=fin:
        millieu=(debut+fin)//2
        if tableau[millieu]==element:
            return True
        elif element<tableau[millieu]:
            fin=millieu-1
        else:
            debut=millieu+1
    return False

print(dicho(29,[14,22,29,35,44,50,56,76]))
print(dicho(61,[14,22,29,35,44,50,56,76]))
```
